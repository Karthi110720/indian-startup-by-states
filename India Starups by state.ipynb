```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('indian_startups_by_state.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='State', palette='viridis')
plt.title('Indian Startups by State', fontweight='bold')
plt.xlabel('State', fontweight='bold')
plt.ylabel('Count', fontweight='bold')
plt.xticks(fontweight='bold', rotation=90)
plt.tight_layout()

plt.show()
```

```python
startup_count_by_year = df.groupby('Year')['Count'].sum()
```

```python
plt.figure(figsize=(10, 6)) 
startup_count_by_year.plot(kind='line', marker='o', linestyle='--', markersize=8) 
plt.ylabel('Number of Startups')
plt.title('Trends in Startup Growth')
plt.show()
```

```python
plt.figure(figsize=(10, 6)) 
startup_count_by_year.plot(kind='line', marker='o', linestyle='--', markersize=8) 
plt.ylabel('Number of Startups')
plt.title('Trends in Startup Growth')
plt.show()
```

```python
df = df.drop('S No.', axis=1)
```

```python
numeric_columns = df.select_dtypes(include=['int64', 'float64'])
```

```python
correlation_matrix = numeric_columns.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, annot_kws={'size': 10})
plt.title('Correlation Matrix')
plt.xticks(rotation=45)
plt.yticks(rotation=0) 
plt.show()
```

```python
startup_count_by_state = df.groupby('State')['Count'].sum()
```

```python
startup_count_by_state = startup_count_by_state.sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
startup_count_by_state.plot(kind='bar', color='red')
plt.xlabel('State')
plt.ylabel('Number of Startups')
plt.title('Distribution of Startups by State')
plt.xticks(rotation=90)
plt.show()
```

```python
startup_count_by_industry = df.groupby('Industry')['Count'].sum()
startup_count_by_industry = startup_count_by_industry.sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
startup_count_by_industry.plot(kind='bar', color='green')
plt.xlabel('Industry')
plt.ylabel('Number of Startups')
plt.title('Distribution of Startups by Industry')
plt.xticks(rotation=90)
plt.show()
```
